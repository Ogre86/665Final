# Data Preparation
Real-world datasets need to be pre-processed (e.g., cleaning, transforming) prior to formal analysis. Perform all necessary data cleaning and transformation activities. If necessary, you need to create new variables from existing variables. See an [example](http://rstudio-pubs-static.s3.amazonaws.com/342210_7c8d57cfdd784cf58dc077d3eb7a2ca3.html).

```python
#remove duplicated rows and keep first occurence
df = df.drop_duplicates(keep="first", inplace=False)
len(df)
```

According to the history of naming these different content ratings, we find M = GP = PG, X = NC-17. 
- We want to replace M and GP with PG, replace X with NC-17, because these two are what we use nowadays. 
- We want to replace “Approved”, “Not Rated”, “Passed”, “Unrated” with the most common rating “R”. 
- We want to replace TV Movie Ratings with their equivalent of MPAA Ratings

```python
replace_values = {'TV-G':'G','TV-Y':'G',
                  'M':'PG','GP':'PG','TV-Y7':'PG','TV-PG':'PG',                  
                  'TV-14':'PG-13',
                  'X':'NC-17',
                  'Approved':'R','Not Rated':'R','Passed':'R','Unrated':'R','TV-MA':'R'
                 }
df = df.replace({'content_rating': replace_values})

#fill null content ratings with most common rating of 'R'
df['content_rating'] = df['content_rating'].fillna(df['content_rating'].describe()['top'])
```

```python
#remove extra spaces in movie title
df["movie_title"].str.rstrip()
```

Based on external analysis of this data set by Yueming Zhang on 12/7/2017 shown [here](http://rstudio-pubs-static.s3.amazonaws.com/342210_7c8d57cfdd784cf58dc077d3eb7a2ca3.html), we can see that: 
- Genre
- Color
- Aspect Ratio
- Language
- Movie Link

are irrelevant to this data set to predict IMDb Score. We will remove them here

```python
df = df.drop(['color','genres','aspect_ratio','language','movie_imdb_link'], axis=1);
```

```python
# this shows the number of columns was successfully dropped
df.shape
```

```python
#replace the null values for average faces with 50% average
df['facenumber_in_poster'] = df['facenumber_in_poster'].fillna(df['facenumber_in_poster'].describe()['50%'])
```

```python
#replace the null values for duration with 50% average
df['duration'] = df['duration'].fillna(df['duration'].describe()['50%'])
```

```python
# fill empty budget and gross with 0
df[['gross', 'budget']] = df[['gross', 'budget']].fillna(value=0)
```

```python
# create a profit column and percentage return on budget
df['profit']= df['gross']-df['budget']
df['return_on_investment_perc']= df['profit']/df['budget']*100
df[['return_on_investment_perc']] = df[['return_on_investment_perc']].fillna(value=0)
```

Here we see almost all the null fields have been filled. The total length of the data frame is 4998 at this point. Becuase the largest amount of missing fields is such a small percentage of the total data, we will drop all nulls becuase it wont make much of a difference when evaluating the data as a whole.

```python
df.info()
```

```python
# drop all nulls
df = df.dropna();
```

# Classification
The goal is to build a classification model to predict if a movie is good or bad. You need to create a new “categorical” column from imdb_score in order to build classification models.
```python
op_labels = ['Bad','Good']
#op_labels = [0,1]
category = [0.,5.9,10.]
dfc['imdb_labels']= pd.cut(dfc['imdb_score'],
labels=op_labels, bins=category, include_lowest=False)
```
